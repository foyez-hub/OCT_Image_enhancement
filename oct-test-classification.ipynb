```python
import os
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array, load_img

# Paths

model_path = '/kaggle/input/superresmodel/superRes_model.h5'
image_folder_path = '/kaggle/input/superrestest/super_resolved_images/output1'
output_folder_path = '/kaggle/working/TestoutputSuperRes'
# Load the saved model
model = load_model(model_path)

# Function to preprocess image
def preprocess_image(image_path, target_size):
    image = load_img(image_path, target_size=target_size)
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    image = image / 255.0
    return image

# Load class labels
class_labels = ['Normal', 'Diabetes', 'Non-Diabetes']  # Update with your actual class names

# Ensure output class folders exist
for class_label in class_labels:
    class_folder_path = os.path.join(output_folder_path, class_label)
    if not os.path.exists(class_folder_path):
        os.makedirs(class_folder_path)

# Recursive function to process images in subfolders
def process_images_in_folder(folder_path):
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        
        if os.path.isdir(item_path):
            # Recursively process subfolders
            process_images_in_folder(item_path)
        elif os.path.isfile(item_path):
            # Process images in current folder
            # Preprocess image to match model input size
            image = preprocess_image(item_path, target_size=(150, 150))  # Update target_size as per your model
            
            # Predict class
            prediction = model.predict(image)
            predicted_class_index = np.argmax(prediction)
            predicted_class = class_labels[predicted_class_index]
            
            # Print prediction probabilities for debug
            print(f'Predicted probabilities: {prediction}')
            
            # Load image using cv2
            image_cv2 = cv2.imread(item_path)
            
            # Write predicted class on image
            cv2.putText(image_cv2, predicted_class, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            
            # Determine output folder based on predicted class
            output_class_folder_path = os.path.join(output_folder_path, predicted_class)
            
            # Create subfolder structure if not exists
            relative_folder_path = os.path.relpath(folder_path, image_folder_path)
            output_subfolder_path = os.path.join(output_class_folder_path, relative_folder_path)
            os.makedirs(output_subfolder_path, exist_ok=True)
            
            # Save image to the output subfolder structure
            output_image_path = os.path.join(output_subfolder_path, item)
            cv2.imwrite(output_image_path, image_cv2)
            
            print(f'Processed {item} - Predicted Class: {predicted_class}')

# Start processing from the top-level image folder
process_images_in_folder(image_folder_path)
```
