```python
import os
import shutil
import random
import zipfile

main_dir = '/kaggle/input/trainsuperres/super_resolved_images'

# Output folders to process
outputs = ['output0', 'output1', 'output2']

# Ratio for train, validation, test splits
train_ratio = 0.7
val_ratio = 0.2
test_ratio = 0.1

# Create directories for train, validation, and test
base_dir = '/kaggle/working'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

for directory in [train_dir, validation_dir, test_dir]:
    for output in outputs:
        os.makedirs(os.path.join(directory, output), exist_ok=True)

# Iterate through each output folder
for output in outputs:
    output_dir = os.path.join(main_dir, output)
    assert os.path.exists(output_dir), f"Directory '{output_dir}' not found."

    # List all subdirectories (e.g., RawDataQA (1), RawDataQA (2), ...)
    subdirs = os.listdir(output_dir)

    # Process each subdirectory (assuming each corresponds to a different class)
    for subdir in subdirs:
        subdir_path = os.path.join(output_dir, subdir)
        if not os.path.isdir(subdir_path):
            print(f"Skipping '{subdir_path}' as it is not a directory.")
            continue
        
        images = [img for img in os.listdir(subdir_path) if img.endswith('.jpg') or img.endswith('.jpeg')]
        num_images = len(images)
        
        if num_images == 0:
            print(f"No images found in directory '{subdir_path}'. Skipping.")
            continue
        
        # Shuffle the images
        random.shuffle(images)
        
        # Calculate split indices
        train_split = int(train_ratio * num_images)
        val_split = int((train_ratio + val_ratio) * num_images)
        
        # Assign images to train, validation, and test directories
        for i, img in enumerate(images):
            src_path = os.path.join(subdir_path, img)
            if i < train_split:
                dst_dir = os.path.join(train_dir, output, subdir)
            elif i < val_split:
                dst_dir = os.path.join(validation_dir, output, subdir)
            else:
                dst_dir = os.path.join(test_dir, output, subdir)
            
            os.makedirs(dst_dir, exist_ok=True)
            
            # Copy only if `src_path` is a file
            if os.path.isfile(src_path):
                dst_path = os.path.join(dst_dir, img)
                shutil.copy(src_path, dst_path)
                print(f"Copied '{src_path}' to '{dst_path}'.")

print("Dataset splitting completed.")

# Create a zip archive of the train, validation, and test directories
zip_filename = '/kaggle/working/dataset.zip'
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, base_dir))

print(f"Dataset zip file saved as '{zip_filename}'")
```
